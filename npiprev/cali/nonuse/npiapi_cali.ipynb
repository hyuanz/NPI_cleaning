https://npiregistry.cms.hhs.gov/api/?number=&enumeration_type=&taxonomy_description=&first_name=&use_first_name_alias=&last_name=&organization_name=&address_purpose=&city=&state=&postal_code=90201&country_code=&limit=10&skip=&pretty=&version=2.1

```python
#max results returned is 200 for each request
#using skip to get more
r = requests.get("""https://npiregistry.cms.hhs.gov/api/?number=&enumeration_type=&taxonomy_description=&first_name=&use_first_name_alias=&last_name=&organization_name=&address_purpose=&city=&state=&postal_code=90201&country_code=&limit=10&skip=&pretty=&version=2.1""")
results_text = r.text
print(results_text)
```

```python
#transfer into json format
results_json = r.json()
```

#### Read API Output is a JSON document, some are nested or contains array

```python
#nested data example
#the below, 'basic' section includes provider information
#there are also other sections such as addresses, taxonomies, identifier, endpoints, etc.
```

```python
print(results_json['results'][0]['basic'])
```

```python
#transfer into table
```

```python
first_record = flatten(results_json['results'][0])
df1 = pd.DataFrame.from_dict(first_record,orient='index').T
df1
```

```python
result_count = results_json['result_count']
print(result_count)
```

#### Output CSV

```python
#loop
```

```python
i = 0

while i < result_count:
    record = flatten(results_json['results'][i])
    # print(record)
    data = pd.DataFrame.from_dict(record,orient='index').T
    if i == 0:
        df = data
    else:
        df = pd.concat([df,data], axis=0, ignore_index=True)
    i+=1
```

```python
pcm = requests.get("""https://npiregistry.cms.hhs.gov/api/?number=&enumeration_type=&taxonomy_description=Prosthetics Case Management&first_name=&use_first_name_alias=&last_name=&organization_name=&address_purpose=&city=&state=CA&postal_code=&country_code=&limit=200&skip=&pretty=&version=2.1""")
pcm_json = pcm.json()
print(pcm_json['results'][0]['basic'])
```

```python
first_pcm = flatten(pcm_json['results'][0])
pcmdf1 = pd.DataFrame.from_dict(first_pcm,orient='index').T
pcmdf1
```

```python
pcm_count = pcm_json['result_count']
print(pcm_count)
```

```python
i = 0
while i < pcm_count:
    record = flatten(pcm_json['results'][i])
    # print(record)
    data = pd.DataFrame.from_dict(record,orient='index').T
    if i == 0:
        pcmdf = data
    else:
        pcmdf = pd.concat([pcmdf,data], axis=0, ignore_index=True)
    i+=1
```
